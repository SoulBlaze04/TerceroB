# Funciones

```python
import string
from collections import Counter

alphabet = string.ascii_lowercase
punctuation = string.punctuation

# Ejercicio 1
def ej1(str):
    s = str
    s = s.lower()

    for letter in alphabet:
        if not(letter in s):
            return False

    return True

# Ejercicio 2
def ej2(str):
    dictionary = {}
    
    for letter in alphabet:
        dictionary[letter] = 0

    s = str
    s = s.replace(" ", "").lower()

    for character in s:
        dictionary[character] += 1

    for c, i in dictionary.items():
        if not(i==1):
            return False

    return True

# Ejercicio 3
def ej3(str1, str2):
    s1 = str1
    s2 = str2

    s1 = s1.lower().replace(" ", "")
    s2 = s2.lower().replace(" ", "")

    for character in punctuation:
        s1 = s1.replace(character, "")
        s2 = s2.replace(character, "")

    if not (len(s1) == len(s2)):
        return False
    else:
        for letter in s1:
            pos = s2.find(letter)
            if pos == -1:
                return False

    return True

# Ejercicio 4
def ej4(r, p):
    ruta = r
    palabra = p

    try:
        texto = open(ruta, 'r')
        res = []
    
        for linea in texto:
            if not ((linea.find(palabra)) == 0):
                res.append(linea)
    
        texto.close()

    except:
        res = "No se encontró el archivo"
    
    finally:
        return res

# Ejercicio 5
def ej5(str):
    s = str
    count = Counter()
    res = []

    for letter in s:
        count[letter] += 1

    for k,v in count.items():
        res.append(v)

    return res

# Ejercicio 6
def ej6(str):

    entrada = str
    counter = 0
    
    for character in entrada:
        if character == "[":
            counter += 1
        elif character == "]":
            counter -= 1
        
        if counter < 0:
            return "NOT OK"
        
    if not (counter == 0):
        return "NOT OK"
    
    return "OK"
```

# Casos de prueba

```python
print("EJERCICIO 1")
print(ej1("The quick brown fox jumps over the lazy dog"))
print(ej1("patata"))
print(ej1("Abcdefghijklmnopqrstuvwxyz"))
print(ej1("The quick brown fox jumps over the 12 lazy dogs"))
print()

print("EJERCICIO 2")
print(ej2("The quick brown fox jumps over the lazy dog"))
print(ej2("qwertyuiopasdfghjklzxcvbnm"))
print()

print("EJERCICIO 3")
print(ej3("Life on mars", "Alien forms"))
print(ej3("Life ;; On m/ar.s", "Al i,en fo.rms"))
print(ej3("Pata", "Patata"))
print(ej3("Pepe", "Juan"))
print(ej3("Life on venus", "Alien forms"))
print()

print("EJERCICIO 4")
print(ej4("prueba", "la"))
print(ej4("patata", "la"))
print()

print("EJERCICIO 5")
print(ej5("abbabcbdbabdbdbabababcbcbab"))
print(ej5("abcde"))
print(ej5("aabbcc"))
print()

print("EJERCICIO 6")
print(ej6("[]"))
print(ej6("[][]"))
print(ej6("[[][]]"))
print(ej6("]["))
print(ej6("][]["))
print(ej6("[]][[]"))
print(ej6("[hola]"))
print(ej6("]hola["))
```
